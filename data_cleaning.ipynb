```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\sachi\Downloads\online_retail_II.xlsx")
```

```python
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
```

MISSING VALUE

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

DUPLICATE ROWS

```python
print("Duplicate Row:",df.duplicated().sum())
```

CLEANING

```python
df = df.drop_duplicates()
df = df.dropna(subset=["Customer ID"])
df=df.dropna(subset=["Description"])
df= df[df["Quantity"]>0]
df = df[df["Price"]>0]
```

NEW COLUMNS

```python
df["Sales"] = df["Quantity"] * df["Price"]
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["MonthName"] = df["InvoiceDate"].dt.month_name()
```

SAVE CLEANED DATASET

```python
df.to_csv("cleaned_online_retail.csv",index=False)
```

```python
print(df.columns.tolist())
```

KPIs

```python
total_revenue = df["Sales"].sum()
total_orders = df["Invoice"].unique()
total_customers = df["Customer ID"].nunique()
total_products = df["StockCode"].nunique()
print("Total Revenue:",round(total_revenue,2))
print("Total Orders:",total_customers)
print("Total Products:",total_products)
```

MONTHLY SALES TREND

```python
monthly_sales = df.groupby(["Year","MonthName"])["Sales"].sum().reset_index()
monthly_sales
```

TOP 10  PRODUCTS

```python
top_products =(
    df.groupby("Description")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_products)
    
```

TOP 10 Customers

```python
top_customers = (
    df.groupby("Customer ID")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_customers)
```

COUNTRY WISE SALES

```python
country_sales =(
    df.groupby("Country")["Sales"]
    .sum()
    .sort_values(ascending=False)
)
print(country_sales)
```

MOST SOLD PRODUCTS

```python
most_sold = (
    df.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(most_sold)
```

AVERAGE ORDER VALUE

```python
average_order_value = (
    df.groupby("Invoice")["Sales"]
               .sum()
               .mean()
)
print("Average Order Value:", round(average_order_value,2))
    
```

SALES BY MONTH

```python
sales_month =(
    df.groupby("MonthName")["Sales"]
    .sum()
    .sort_values(ascending=False)
)
print(sales_month)
    
```

CANCELLED ORDERS

```python
cancelled = df[df["Invoice"].astype(str).str.startswith("C")]
print("Cancelled Order:",cancelled["Invoice"].nunique())
```

EXPORT SUMMARY

```python
country_sales.to_csv("country_sales.csv")
top_products.to_csv("top_products.csv")
top_customers.to_csv("top_customers.csv")
monthly_sales.to_csv("monthly_sales.csv")
```
